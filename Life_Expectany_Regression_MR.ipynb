# 데이터 불러오기

* 0   Country : 나라 이름 (193 개국)
* 1   Year    : 조사 연도
* 2   Status  : Developed or Developing (개발 도상국 / 선진국)
* 3   Life expectancy : 기대 수명
* 4   Adult Mortality : 인구 1000명 당 15세에서 60세 사이의 사망률 (남성, 여성 포함)
* 5   infant deaths   : 인구 1000명 당 유아 사망 수
* 6   Alcohol         : 15세 이상의 1인당 알콜 소비량(liter)
* 7   percentage expenditure : 1인당 국내총생산(GDP) 대비 보건 예산 지출 비율 (%)
* 8   Hepatitis B : 1세 사이의 B형 간염(HepB) 예방접종 보급률 (%)
* 9   Measles     : 홍역, 인구 1000명당 보고된 환자 수
* 10  BMI         : 전체 인구의 평균 체질량 지수(BMI)
* 11  under-five deaths : 인구 1000명당 5세 미만 사망자 수
* 12  Polio             : 1세 사이의 소아마비(Pol3) 예방접종 보급률 (%)
* 13  Total expenditure : 보건에 대한 정부 지출 비율 (총 정부 지출 중) (%)
* 14  Diphtheria: 1세 사이의 디프테리아(Diphtheria) 파상풍(tetanus) 변독소(toxoid) and 백일해(pertussis) (DTP3) 에방접종 보급률 (%) 
* 15  HIV/AIDS: 신생아 1000명당 사망률 HIV/AIDS (0-4 years) 
* 16  GDP: 1인당 국내총생산 (in USD) 
* 17  Population : 인구수
* 18  thinness  1-19 years : 10세에서 19세 사이의 청소년 마름(thinness) 유병율 (%)
* 19  thinness 5-9 years : 5세에서 9세 사이의 아동 마름(thinness) 유병율 (%)
* 20  Income composition of resources : 자원의 소득 구성에 따른 인간개발지수(Human Development Index) (index ranging from 0 to 1)
* 21  Schooling : 학교 교육 년수 (years)

### 예측

```python
label_name = 'Life_Expectancy'
```

```python
feature_names = df_org.columns.tolist()
feature_names.remove(label_name)
feature_names
```

```python
X = df_org[feature_names]
y = df_org[label_name]

X.shape, y.shape
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,shuffle=True, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=42)
model
```

```python
from sklearn.model_selection import cross_val_predict

y_valid_predict = cross_val_predict(model, X_train, y_train, cv=3)
len(y_valid_predict)
```

```python
from sklearn.model_selection import RandomizedSearchCV
param_distributions = {"n_estimators" : np.random.randint(0, 1000, 200), 
                       "max_depth" : np.random.randint(0, 200, 150)
                      }

reg = RandomizedSearchCV(model, 
                         param_distributions=param_distributions, 
                         n_iter=5, cv=5, n_jobs=-1)
reg.fit(X_train, y_train)
```

```python
reg.best_estimator_
```

```python
reg.best_score_
```

```python
pd.DataFrame(reg.cv_results_)
```

```python
y_predict = reg.fit(X_train, y_train).predict(X_test)
```
