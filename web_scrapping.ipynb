```python
def market_within_10days(m_symbol): # make all above to function
    url = f'https://finance.naver.com/world/sise.naver?symbol={m_symbol}'
    headers = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.54 Safari/537.36"}

    response = requests.get(url, headers = headers)
    html = bs(response.text, 'lxml')
    tables = html.select('table')
    table = pd.read_html(str(tables),encoding='cp949')
    df_table = table[1]
    return df_table
```

```python
m_symbol = 'NAS@IXIC' # Check Nasdaq Composite
market_within_10days(m_symbol)
```
